```python
from google.colab import drive
drive.mount('/content/drive')
import sys
import pandas as pd
sys.path.append("/content/drive/MyDrive/Colab Notebooks/my_lib")
import micromlgen
from sklearn.ensemble import RandomForestRegressor

# Membaca file CSV
df = pd.read_csv('Jahe kering mix 1.csv', header=0)
names = df.columns.tolist()
# Memisahkan kolom menjadi target dan fitur
y = df.iloc[:, 0]  # target
X = df.iloc[:,1:]  # fitur

# 1. Buat model sederhana
clf = RandomForestRegressor(n_estimators=100, min_samples_leaf=1, max_depth=20, max_features=1, ).fit(X, y)

# 2. Coba konversi menggunakan micromlgen
try:
    c_code = micromlgen.port(clf)
    print("Berhasil! Library berfungsi dengan baik.")
    print("Contoh potongan kode C++ hasil konversi:")
    print(c_code[:150], "...")
except AttributeError:
    print("Library terdeteksi, tapi fungsi 'port' tidak ditemukan. Cek dokumentasi library Anda.")
except Exception as e:
    print(f"Terjadi error lain: {e}")
```

```python
# Asumsikan c_code adalah variabel hasil konversi micromlgen.port(clf)
c_code = micromlgen.port(clf)

# Proses simpan file
file_name = "c_RFjahe.h"
with open(file_name, "w") as f:
    f.write(c_code)

print(f"File berhasil disimpan sebagai {file_name}")
```
